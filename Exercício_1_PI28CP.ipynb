```python
import cv2 as cv
import numpy as np
from matplotlib import image
import matplotlib.pyplot as plt

def show_image(img: np.ndarray, title: str = "", x_step: int = 1, y_step: int = 1):
    plt.figure(figsize=(6, 6))
    plt.xticks(np.arange(0, img.shape[1], x_step))
    plt.yticks(np.arange(0, img.shape[0], y_step))
    plt.xlim(-0.5, img.shape[1] - 0.5)  # centralizar os pixels no eixo x
    plt.ylim(img.shape[0] - 0.5, -0.5)  # centralizar os pixels no eixo y
    plt.title(title)
    plt.imshow(img)
    plt.show()
```

## h) Recarregue a imagem original

```python
test_img = np.array(image.imread("./img/test.jpeg"))
show_image(test_img, "Imagem", 100, 110)
```

## i) Separe os canais da imagem

```python
R, G, B = test_img[:, :, 0], test_img[:, :, 1], test_img[:, :, 2]
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
im = axs[0].imshow(R, vmin=0, vmax=255)
axs[0].set_title("Canal Vermelho")
cbar = fig.colorbar(im, ax=axs[0])
cbar.set_ticks([0, 127, 255])

im = axs[1].imshow(G, vmin=0, vmax=255)
axs[1].set_title("Canal Verde")
cbar = fig.colorbar(im, ax=axs[1])
cbar.set_ticks([0, 127, 255])

im = axs[2].imshow(B, vmin=0, vmax=255)
axs[2].set_title("Canal Azul")
cbar = fig.colorbar(im, ax=axs[2])
cbar.set_ticks([0, 127, 255])
plt.show()
```

## j) Calcule o resto da divisão de cada pixel do canal vermelho por 128

```python
print(R)
R %= 128
print(R)
```

## k) Some cada pixel do canal azul por 10

```python
print(B)
B = B + 10
print(B)
```

## l) Calcule o módulo da subtração de pixel do canal verde por 15

## m) Plote e salve a imagem resultante das operações

```python
RGB = np.stack((R, G, B), axis=-1)
show_image(RGB, "Teste", 100, 110)
```

## o) Desenhe uma linha, um círculo e um retângulo na imagem

```python
height, width, _ = test_img.shape
center = (width // 2, height // 2)
circle_radius = width // 2
cv.circle(test_img, center, circle_radius, (0, 0, 0), 4)
rectangle_top_left = (center[0] - circle_radius // 2, center[1] - circle_radius // 2)
rectangle_bottom_right = (center[0] + circle_radius // 2, center[1] + circle_radius // 2)
cv.rectangle(test_img, rectangle_top_left, rectangle_bottom_right, (0, 0, 0), 5)
cv.line(test_img, (0, 0), (width, height), (0, 0, 0), 6)
```
