```python
import cv2 as cv
import numpy as np
from matplotlib import image
import matplotlib.pyplot as plt

def show_image(img: np.ndarray, title: str = ""):
    plt.title(title)
    plt.imshow(img)
    plt.show()
    plt.close()

img1 = np.array(image.imread("../img/img1.jpeg"))
img2 = np.array(image.imread("../img/img2.png"))
print(img1.shape, img2.shape)

show_image(img1, "Cenoura")
show_image(img2, "Pessoa")
```

a) Realize a translação de 6 pontos em X e 6 pontos em Y


```python
h1, w1, _ = img1.shape
h2, w2, _ = img2.shape

M = np.float32([[1, 0, 6],[0, 1, 6]])
img1_tr = cv.warpAffine(img1, M, (w1, h1))
img2_tr = cv.warpAffine(img2, M, (w2, h2))

print(img1.shape, img1_tr.shape)
print(img2_tr.shape, img2.shape)

show_image(img1_tr, "Cenoura")
show_image(img2_tr, "Pessoa")
```
